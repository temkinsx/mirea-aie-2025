## 1. Импорт библиотеки NumPy

```python
import numpy as np
```

## 2. Примеры создания массивов

### 2.1. Из обычных списков

```python
# из списка
a = np.array([0, 142, 3453, 3242])
print("Вектор из обычного списка: ", a, "\nТип: ", type(a), "\nФорма: ", a.shape)
```

```python
# из двумерного списка
b = np.array([[1, 2, 3, 4], [5, 6, 7, 8]])
print("Двумерный массив из двумерного списка: \n", b, "\nТип: ", type(b), "\nФорма: ", b.shape)
```

### 2.2. С использованием функций из библиотеки NumPy

#### 2.2.1. np.zeros
np.zeros возвращает массив заданной формы, заполненный нулями

```python
a = np.zeros((5, 5))
print("Двумерный массив 5 x 5, заполненный нулями: \n", a, "\nТип: ", type(a), "\nФорма: ", a.shape)
```

#### 2.2.2. np.ones
np.ones возвращает массив заданной формы, заполенный единицами

```python
a = np.ones((3, 3, 3))
print("Трехмерный массив 3 x 3 x 3, заполенный единицами: ", a, "\nТип: ", type(a), "\nФорма: ", a.shape)
```

#### 2.2.3 np.arange
np.arange возвращает массив из последовательности чисел по заданному началу, концу и шагу

```python
a = np.arange(1, 50, 2)
print("Массив из последовательности чисел от 1 до 50 с шагом 2: ", a, "\nТип: ", type(a), "\nФорма: ", a.shape)
```

```python
a = np.array([np.arange(1, 50, 2), np.arange(1, 50, 2)])
print(a)
```

#### 2.2.4 np.reshape
np.reshape изменяет форму массива, не меняя данные в нем
- число элементов должно совпадать
- один из размеров можно оставить равным -1, тогда NumPy сам вычислит его

```python
a = np.arange(-25, 25).reshape (5, 10)
b = np.arange(0, 30).reshape(2, 3, -1)
print("Двумерный массив 5 x 10 из последовательности чисел от -25 до 24: ", a)
print("Трехмерный массив 2 x 3 x 5 из последовательности чисел от 0 до 29: ", b)
```

## 3. Индексирование и срезы

### 3.1 Получение и изменение отдельного элемента

```python
a = np.arange(1, 6) ** 2
print("Массив: ", a)
print("Третий элемент массива: ", a[2])
a[2] = 500
print("Массив после изменения третьего элемента: ", a)
```

### 3.2 Получение и изменение среза элементов массива
NumPy срез - "окно" в массив, от которого он взят (т.е. при изменении элемента в NumPy срезе это отразится на оригинальном массиве)

```python
a = np.arange(1, 10, 2)
print("Массив: ", a)
b = a[0:3]
print("Срез:", b)
a[0:3] = a[0:3] ** 2
print("Массив после изменения элементов с индексами от 0 до 3", a)
b[0] = 12345
print("Срез после изменения первого элемента: ", b)
print("Массив после изменения элемента в срезе: ", a)
```

### 3.3 Разворот массива

```python
a = np.arange(1, 10)
print("Массив: ", a)
print("Развернутый массив: ", a[::-1])
```

### 3.4 Итерирование по массиву

```python
a = np.arange(20).reshape(4, 5)
print("Массив :\n", a)
for el in a.flat:
    print(el)
```

## 4. Векторные операции и базовые статистики

### 4.1. Агрегирующие функции

```python
a = np.array((np.arange(10, 19), np.arange(-9, 0), np.arange(1, 10)))
print("Массив: \n", a)
print("Сумма элементов: ", np.sum(a))
print("Минимальный элемент: ", a.min())
print("Максимальный элемент: ", a.max())
print("Минимумы каждой строки: ", a.min(axis=1))
print("Максимумы каждой строки: ", a.max(axis=1))
print("Среднее: ", a.mean())
```

### 4.2. Поэлементное сложение и умножение

```python
a = np.array([[1, 2], [3, 4]])
print("Первый массив: \n", a)
b = np.array([[10, 20], [30, 40]])
print("Второй массив : \n", b)

print("Поэлементная сумма: \n", a + b)
print("Поэлементное умножение: \n", a * b)
```
